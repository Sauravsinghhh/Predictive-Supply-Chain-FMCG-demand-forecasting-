# 01. Data Ingestion & Schema Validation Demo

This notebook demonstrates loading, memory optimization, and schema validation for the **FreshMind: FMCG demand forecasting** project using our modular `DataLoader` pipeline.

## 1. Setup Environment and Path Imports

```python
import os
import sys
import pandas as pd

# Add root folder to path so we can import src module
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))

from src.data.data_loader import DataLoader
from src.utils.logging_utils import setup_logger

print("Environment setup completed successfully.")
```

## 2. Initialize DataLoader

```python
# Load configurations from relative path configs/config.yaml
loader = DataLoader(config_path="../configs/config.yaml")
```

## 3. Run Ingestion and Validation checks

We will run the validation checks, which will downcast data types (saving memory), check schemas, look for duplicate/null records, and save a data status markdown report under `reports/data_status_report.md`.

```python
validation_results = loader.run_validation_checks()
print("Validation pipeline run complete.")
```

## 4. Inspect Results Summary

```python
for key in ["calendar", "sell_prices", "sales"]:
    info = validation_results[key]
    print(f"=== Dataset: {key.upper()} ===")
    if "shape" in info:
        print(f"Shape: {info['shape']}")
    else:
        print(f"Shape: ({info['shape']}, {info['columns_count']} columns)")
    print(f"Schema Pass: {info['schema_check_pass']}")
    print(f"Null Records Count: {info['null_count']}")
    print(f"Duplicate Rows: {info['duplicate_count']}")
    if "invalid_prices_count" in info:
        print(f"Invalid Price Count (<= 0): {info['invalid_prices_count']}")
    if "negative_sales_count" in info:
        print(f"Negative Sales Volumes Count: {info['negative_sales_count']}")
    print("\n")
```

## 5. Execute Sample Query

Confirming that we can easily load, query, and structure the data after validation. Let's see the filtered sales data for `HOBBIES_1_001` in `CA_1`:

```python
print(f"Query: {validation_results['sample_query']['query_description']}")
print(f"Rows returned: {validation_results['sample_query']['rows_returned']}")

sample_df = pd.DataFrame(validation_results['sample_query']['sample_data'])
sample_df
```
